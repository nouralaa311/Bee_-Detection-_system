```python
from ultralytics import YOLO

model = YOLO('yolo11m.pt')
results = model.train(
    data='/content/Queen-Bee-4/data.yaml',
    epochs=120,
    imgsz=640,
    batch=8,
    name='yolo11m_bee',
    device=0,
    patience=25,
    amp=True,
    exist_ok=True,
    plots=True,
    optimizer='AdamW',
    lr0=0.0007,
    lrf=0.01,
    momentum=0.9,
    weight_decay=0.0005,
    warmup_epochs=2.0,
    warmup_momentum=0.8,
    warmup_bias_lr=0.1
)
```

```python
import os
from google.colab import files

best = 'runs/detect/yolo11m_bee/weights/best.pt'

if os.path.exists(best):
    files.download(best)
    print("🎉 Downloaded!")
else:
    import glob
    models = glob.glob('runs/detect/*/weights/best.pt')
    print("Available:", models)
    if models:
        files.download(models[-1])
    else:
        print("No 'best.pt' model found in any 'runs/detect' directory.")
```

```python
metrics = model.val()
print(f"mAP50-95: {metrics.box.map}")
print(f"mAP50: {metrics.box.map50}")
print(f"mAP75: {metrics.box.map75}")
print(f"Precision: {metrics.box.mp}")
print(f"Recall: {metrics.box.mr}")

print(f"mAP50-95 per class: {metrics.box.maps}")
```
